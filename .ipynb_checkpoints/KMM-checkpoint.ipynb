```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

from KMM_functions import kmm_weights, spatial_kmm_cv 
```

```python
np.random.seed(42)
n_samples = 300

X_coords = np.random.uniform(-5, 5, size=(n_samples, 2))
y = (np.sin(X_coords[:, 0]) * 2 + np.cos(X_coords[:, 1]) * 1.5 +
     np.random.normal(0, 0.3, n_samples))

plt.scatter(X_coords[:, 0], X_coords[:, 1], c=y, cmap='viridis')
plt.colorbar(label='y')
plt.show()

mse_kmm, mse_unw = spatial_kmm_cv(X_coords, y, n_folds=5, gamma=0.5, B=10)
print(f"MSE с KMM‑взвешиванием: {mse_kmm:.4f}")
print(f"MSE без взвешивания (обычная пространственная CV): {mse_unw:.4f}")

print("\nСтандартная случайная 5‑fold CV (для ориентира):")
kf = KFold(n_splits=5, shuffle=True, random_state=42)
errors_random = []
for train_idx, test_idx in kf.split(X_coords):
    X_tr, X_te = X_coords[train_idx], X_coords[test_idx]
    y_tr, y_te = y[train_idx], y[test_idx]
    model = LinearRegression()
    model.fit(X_tr, y_tr)
    y_pred = model.predict(X_te)
    errors_random.append(mean_squared_error(y_te, y_pred))
print(f"MSE (случайная CV): {np.mean(errors_random):.4f}")
```
